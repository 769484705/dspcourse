加载画图和科学计算包

```python

import matplotlib.pyplot as plt
import numpy as np
```

产生$0.1\sim 2\pi$的正弦信号和幅度为1的常数信号

```python

x = np.linspace(0.1, 2 * np.pi, 41)
y1 = np.sin(x)
y2 = np.cos(2*x)
```

##1. 相加、减运算

```python
y  = y1 + y2
yy = y1 - y2
plt.subplot(121)
plt.stem(x, y)
plt.subplot(122)
plt.stem(x, yy)
plt.show()
```

## 2. 相乘、除

```python
y  = y1 * y2 # element-wise multiplication
yy = y1 / y2 # element-wise division
plt.subplot(121)
plt.stem(x, y)
plt.subplot(122)
plt.stem(x, yy)
plt.show()
```

## 4. 序列的平移和翻转

循环平移函数

```python
def cshift(key,array):
    return np.append(array[-key:],array[:-key])
```

线性平移

```python
def lshift(key,array):
    if key<0:
        return np.append(array[-key:],np.zeros([-key]))
    else:
        return np.append(np.zeros([key]),array[:-key])
```

```python
ylr = np.flipud(y1)     # flip left and right
yls = lshift(-10,y1)    # left shift
yrs = lshift(10,y1)     # right shift 
```

```python
plt.subplot(221)
plt.title('Original')
plt.stem(x, y1)
plt.subplot(222)
plt.title('Reversal')
plt.stem(x, ylr)
plt.subplot(223)
plt.title('Linear Shift -10')
plt.stem(x, yls)
plt.subplot(224)
plt.title('Linear Shift 10')
plt.stem(x, yrs)
plt.show()
```
