那么，本节我们将简单给出一个Python产生一个离散正弦序列的例子，并显示出来。

```python
import matplotlib.pyplot as plt
import numpy as np
 
x = np.linspace(0.1, 2 * np.pi, 41)
y = np.sin(x)
 
plt.stem(x, y)
plt.show()

plt.plot(x,y)
```
